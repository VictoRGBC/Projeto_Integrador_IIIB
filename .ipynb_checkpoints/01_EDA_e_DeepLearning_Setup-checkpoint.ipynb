```python
# Célula 1: Carregar Bibliotecas e Dados

import pandas as pd

# Ajuste o nome do arquivo, separador (sep), decimal e adicione a codificação!
caminho_arquivo = 'STP-20251124190950218.csv' 
df = pd.read_csv(caminho_arquivo, sep=';', decimal=',', encoding='latin-1') 

# ----------------------------------------------------
# Inspecionar o DataFrame

print("Nomes das colunas:")
print(df.columns)
print("-" * 30)
print("Primeiras 5 linhas (Formato dos dados):")
print(df.head())
```

```python
## Célula de Tratamento de Dados (Para Séries Temporais)

# O nome da coluna longa que identificamos
NOME_COLUNA_LONGA = '7465 - Índice de Preços ao Consumidor (IPC-Fipe) - Alimentação - Var. % mensal'
NOME_COLUNA_ALVO = 'Var_IPC_Alimentacao_Mensal'

# Passo 1A: Renomear, Corrigir e Converter a Variável Alvo para PODER LIMPAR

# Renomear (Para facilitar o acesso)
df.rename(columns={NOME_COLUNA_LONGA: NOME_COLUNA_ALVO}, inplace=True)

# Corrigir o formato: Substituir a vírgula (,) por ponto (.)
df[NOME_COLUNA_ALVO] = df[NOME_COLUNA_ALVO].astype(str).str.replace(',', '.', regex=False)

# Passo 1B: Limpar dados ruins (como a linha "Fonte")
# Identificamos e removemos linhas onde a coluna ALVO não é um número válido (texto, notas de rodapé, etc.)
valores_numericos = pd.to_numeric(df[NOME_COLUNA_ALVO], errors='coerce') 
df_limpo = df[valores_numericos.notna()].copy() 

# ----------------------------------------------------

# Passo 2: Agora que o DataFrame está limpo, podemos tratar a data.
df_limpo['Data'] = pd.to_datetime(df_limpo['Data'], format='%m/%Y')
df_limpo.set_index('Data', inplace=True)
df_ts = df_limpo.sort_index()

# Passo 3: Conversão final para float e remoção de NaNs remanescentes
df_ts[NOME_COLUNA_ALVO] = pd.to_numeric(df_ts[NOME_COLUNA_ALVO]) # Conversão final
df_final = df_ts[[NOME_COLUNA_ALVO]].dropna()

# Passo 4: Salvar o Arquivo Limpo
df_final.to_csv('dados_inflacao_corrigidos.csv', index=True, sep=',', decimal='.')

print("Tratamento de Dados Finalizado.")
print("Série Temporal (IPC Alimentação) pronta para modelagem:")
print(df_final.head())
print(df_final.info())
```

```python
# 1. Renomear a coluna longa para um nome simples
NOME_COLUNA_ALVO = 'Var_IPC_Alimentacao_Mensal'
df_ts.rename(columns={df_ts.columns[0]: NOME_COLUNA_ALVO}, inplace=True)

# 2. Corrigir o formato: Substituir a vírgula (,) por ponto (.) e converter para float
# Se a leitura inicial com 'decimal=,' não converteu, fazemos explicitamente agora.
df_ts[NOME_COLUNA_ALVO] = df_ts[NOME_COLUNA_ALVO].astype(str).str.replace(',', '.', regex=False)
df_ts[NOME_COLUNA_ALVO] = pd.to_numeric(df_ts[NOME_COLUNA_ALVO])

# Selecionar apenas a série temporal final e remover NaNs
df_final = df_ts[[NOME_COLUNA_ALVO]].dropna()
```

```python
# Salvar o DataFrame final pronto para modelagem
# Usando separador (sep) e decimal padrão para ML (vírgula e ponto)
df_final.to_csv('dados_inflacao_corrigidos.csv', index=True, sep=',', decimal='.')

print("Tratamento de Dados Finalizado.")
print("Série Temporal (IPC Alimentação) pronta para modelagem:")
print(df_final.head())
print(df_final.info())
```

```python
## Célula 3: Preparação dos Dados para LSTM

# Importar bibliotecas necessárias para Deep Learning
from sklearn.preprocessing import MinMaxScaler
import numpy as np
# Note: train_test_split do sklearn não é ideal para séries temporais, 
# por isso usamos a divisão manual X[:-test_size], que já está no código.

# 1. Definir o tamanho da janela de tempo (Lookback period)
# Usaremos os últimos 12 meses (1 ano) para prever o próximo mês.
look_back = 12 

# 2. Normalizar os dados (escalonar entre 0 e 1)
# O df_final foi criado na célula anterior.
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(df_final.values)

# 3. Função para criar as janelas de tempo (Sequences)
# 
def create_sequences(data, look_back):
    X, Y = [], []
    for i in range(len(data) - look_back):
        # X: Janela de dados passados (12 meses)
        a = data[i:(i + look_back), 0]
        X.append(a)
        # Y: Valor futuro a ser previsto (o mês seguinte)
        Y.append(data[i + look_back, 0])
    return np.array(X), np.array(Y)

# 4. Criar as sequências X (Entrada) e Y (Saída)
X, Y = create_sequences(data_scaled, look_back)

# 5. Dividir em Treino e Teste
# Usaremos os últimos 5 anos (60 meses) para teste (test_size = 60)
test_size = 60
X_train, X_test = X[:-test_size], X[-test_size:]
Y_train, Y_test = Y[:-test_size], Y[-test_size:]

# 6. Remodelar os dados para o formato LSTM [amostras, passos de tempo, features]
# (X_train.shape[0], X_train.shape[1], 1)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

print(f"Formato X de Treino (Janelas): {X_train.shape}")
print(f"Formato X de Teste (Janelas): {X_test.shape}")
print(f"Total de Amostras de Treino: {X_train.shape[0]}")
```

```python
## Célula 4: Construção e Treinamento do Modelo LSTM

# Importar Keras (TensorFlow)
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 1. Definição da Arquitetura do Modelo
model = Sequential()
# Camada LSTM: usa 50 unidades de memória, retorna sequências (return_sequences=True) 
# para a próxima camada LSTM e define o formato de entrada.
model.add(LSTM(units=50, 
               return_sequences=True, 
               input_shape=(X_train.shape[1], 1))) 
model.add(Dropout(0.2)) # Camada para evitar Overfitting

# Segunda Camada LSTM (para maior complexidade)
model.add(LSTM(units=50, return_sequences=False)) # return_sequences=False na última camada LSTM
model.add(Dropout(0.2)) 

# Camada Densa (Output Layer): Apenas 1 unidade, pois estamos prevendo 1 valor (inflação do mês seguinte).
model.add(Dense(units=1)) 

# 2. Compilação do Modelo
# Otimizador: 'adam' (padrão e robusto)
# Função de Perda: 'mean_squared_error' (MSE), padrão para problemas de regressão/previsão de valores.
model.compile(optimizer='adam', loss='mean_squared_error')

# Exibir o resumo do modelo
print("--- Resumo da Arquitetura LSTM ---")
model.summary()
print("---------------------------------")

# 3. Treinamento do Modelo
print("Iniciando o Treinamento...")
history = model.fit(X_train, 
                    Y_train, 
                    epochs=100,      # Número de iterações (Epochs). Um número maior pode ser melhor.
                    batch_size=32,   # Quantidade de amostras processadas por vez.
                    validation_data=(X_test, Y_test), # Usamos o teste para monitorar a perda
                    verbose=1)

print("Treinamento concluído!")
```

```python
## Célula 5: Avaliação do Modelo e Predição

from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np # Garante que numpy esteja importado

# 1. Fazer Previsões
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# 2. Desnormalizar as Previsões e os Valores Reais
# Reverte a transformação para a escala de inflação original (o 'scaler' foi criado na Célula 3)
train_predict = scaler.inverse_transform(train_predict)
Y_train_inverse = scaler.inverse_transform(Y_train.reshape(-1, 1)) 

test_predict = scaler.inverse_transform(test_predict)
Y_test_inverse = scaler.inverse_transform(Y_test.reshape(-1, 1))

# 3. Cálculo da Métrica de Desempenho (RMSE)
# Erro Quadrático Médio da Raiz (Root Mean Squared Error) - Métrica padrão para modelos de regressão.
train_rmse = sqrt(mean_squared_error(Y_train_inverse, train_predict))
test_rmse = sqrt(mean_squared_error(Y_test_inverse, test_predict))

print("--- Métricas de Erro (Valores Reais) ---")
print(f"RMSE de Treino: {train_rmse:.4f}")
print(f"RMSE de Teste: {test_rmse:.4f}")
print("----------------------------------------")

# 4. Preparação para Visualização
# Criar um DataFrame com a série original
df_pred_full = df_final.copy()
df_pred_full['Previsao_LSTM'] = np.nan # Inicializa coluna de previsão com NaN

# Posições de início e fim no DataFrame original
# Subtrai o tamanho da janela (look_back=12) e o número de testes (test_size=60) do total
start_test_index = len(df_final) - len(Y_test_inverse)

# Inserir previsões no DF original (apenas para o período de teste)
df_pred_full.iloc[start_test_index:, df_pred_full.columns.get_loc('Previsao_LSTM')] = test_predict.flatten()

# 5. Plotar Resultados 
plt.figure(figsize=(14, 7))
plt.plot(df_pred_full.index, df_pred_full['Var_IPC_Alimentacao_Mensal'], label='Valores Reais (IPC-Fipe)')
plt.plot(df_pred_full.index, df_pred_full['Previsao_LSTM'], label='Previsão LSTM (Teste)', color='red', linestyle='--')
plt.title('Previsão de Inflação (IPC-Fipe - LSTM)')
plt.xlabel('Data')
plt.ylabel('Variação Mensal (%)')
plt.legend()
plt.grid(True)
plt.show()
```
